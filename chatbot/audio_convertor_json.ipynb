```python
from gtts import gTTS
from pydub import AudioSegment
import base64
import io

# Step 1: Generate speech using gTTS (Google Text-to-Speech)
text = "Give me a fat loss plan"
tts = gTTS(text, lang='en')
tts.save("output.mp3")

# Step 2: Convert MP3 to WAV (16-bit PCM, 16kHz)
audio = AudioSegment.from_mp3(r"F:\quest_digiflex\output.mp3")
audio = audio.set_frame_rate(16000).set_channels(1).set_sample_width(2)  # 16kHz, mono, 16-bit
audio.export("output.wav", format="wav")

# Step 3: Encode WAV file to Base64
with open("output.wav", "rb") as f:
    audio_bytes = f.read()
    encoded_audio = base64.b64encode(audio_bytes).decode("utf-8")

# Step 4: Print or save the JSON body to paste in Postman
json_body = {
    "audio": encoded_audio,
    "language": "en-US"
}

print("Use this JSON in Postman:\n")
import json
print(json.dumps(json_body, indent=2))
```
